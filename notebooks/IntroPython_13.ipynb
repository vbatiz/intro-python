## ¿Qué es Plotly Express?

**Plotly Express** (`plotly.express`, abreviado como `px`) es una biblioteca de visualización de datos de alto nivel para Python. Permite crear gráficas interactivas y atractivas con muy pocas líneas de código. Puede revisar la documentación [aquí](https://plotly.com/python/plotly-express/).

### ¿Por qué usar Plotly Express?
- Gráficas **interactivas** (zoom, tooltips, filtros)
- Sintaxis **sencilla y consistente**
- Compatible con **pandas DataFrames**
- Gran variedad de tipos de gráficas
- Exportable a HTML, PNG, PDF

### Temas de este notebook

| # | Tipo de Gráfica | Uso recomendado |
|---|---|---|
| 1 | Gráfica de dispersión | Relación entre variables numéricas |
| 2 | Gráfica de líneas | Tendencias en el tiempo |
| 3 | Gráfica de barras | Comparación de categorías |
| 4 | Histograma | Distribución de una variable |
| 5 | Box Plot | Distribución y valores atípicos |
| 6 | Violin Plot | Distribución detallada por grupo |
| 7 | Gráfica de pastel | Proporciones de un todo |
| 8 | Mapa de calor (Heatmap) | Correlaciones y matrices |
| 9 | Gráfica de área | Evolución acumulada |
| 10 | Scatter Matrix | Relaciones múltiples |
| 11 | Gráfica 3D | Tres variables simultáneas |
| 12 | Treemap | Jerarquías proporcionales |
| 13 | Sunburst | Jerarquías circulares |
| 14 | Mapas geográficos | Datos con coordenadas |
| 15 | Animaciones | Datos a lo largo del tiempo |
| 16 | Facets (subgráficas) | Comparar por categorías |
| 17 | Personalización | Temas y estilos |

## 1 - Gráfica de Dispersión (Scatter Plot)

### ¿Qué es?
La gráfica de dispersión muestra la relación entre **dos variables numéricas**. Cada punto representa una observación. Es fundamental para:
- Detectar **correlaciones** entre variables
- Identificar **valores atípicos** (outliers)
- Explorar **patrones** en los datos

### Parámetros clave
| Parámetro | Descripción |
|---|---|
| `x` | Variable del eje horizontal |
| `y` | Variable del eje vertical |
| `color` | Colorea los puntos según una variable |
| `size` | Tamaño de los puntos según una variable |
| `hover_name` | Etiqueta al pasar el mouse |
| `facet_col` | Divide en sub-gráficas por columnas |

````python
# Gráfica de dispersión básica
# Relación entre PIB per cápita y esperanza de vida por país

df_2007 = df_gapminder[df_gapminder['year'] == 2007]

fig = px.scatter(
    df_2007,
    x='gdpPercap',          # Eje X: PIB per cápita
    y='lifeExp',            # Eje Y: Esperanza de vida
    color='continent',      # Color por continente
    size='pop',             # Tamaño por población
    hover_name='country',   # Al pasar el mouse muestra el país
    size_max=60,            # Tamaño máximo de los puntos
    log_x=True,             # Escala logarítmica en X
    title='PIB per cápita vs Esperanza de Vida (2007)',
    labels={
        'gdpPercap': 'PIB per cápita (USD, escala log)',
        'lifeExp': 'Esperanza de Vida (años)',
        'continent': 'Continente',
        'pop': 'Población'
    }
)

fig.show()
print('\nPista: Pasa el mouse sobre los puntos para ver el nombre del país.')
print('   Haz zoom arrastrando el mouse. Haz doble clic para restablecer.')
````

---
## 2 - Gráfica de Líneas (Line Chart)

### ¿Qué es?
Ideal para visualizar la **evolución de una o más variables a lo largo del tiempo**. Conecta puntos de datos con líneas para mostrar tendencias y patrones temporales.

### Cuándo usarla
- Series de tiempo (precios, temperatura, ventas)
- Comparar tendencias entre grupos
- Detectar estacionalidad o ciclos

````python
# Evolución de acciones tecnológicas
fig = px.line(
    df_stocks,
    x='date',
    y=['GOOG', 'AAPL', 'AMZN', 'MSFT'],   # Múltiples líneas
    title='Evolución del precio de acciones tecnológicas',
    labels={'value': 'Precio (USD)', 'date': 'Fecha', 'variable': 'Empresa'},
    markers=True   # Puntos en cada observación
)

fig.update_layout(
    hovermode='x unified'   # Tooltip unificado al pasar el mouse
)

fig.show()

# --- Segundo ejemplo: Esperanza de vida de países seleccionados ---
paises = ['Mexico', 'Brazil', 'Argentina', 'Colombia', 'Chile']
df_latam = df_gapminder[df_gapminder['country'].isin(paises)]

fig2 = px.line(
    df_latam,
    x='year',
    y='lifeExp',
    color='country',
    markers=True,
    title='Esperanza de vida en Latinoamérica (1952–2007)',
    labels={'lifeExp': 'Esperanza de Vida (años)', 'year': 'Año', 'country': 'País'}
)

fig2.show()
````

---
## 3 - Gráfica de Barras (Bar Chart)

### ¿Qué es?
Compara valores entre **categorías discretas**. Es la gráfica más común para visualizar conteos, sumas o promedios agrupados.

### Variantes principales
- `barmode='group'` → Barras agrupadas lado a lado
- `barmode='stack'` → Barras apiladas (muestra composición)
- `orientation='h'` → Barras horizontales

````python
# Barras agrupadas: Propinas promedio por día y género
df_tips_agg = df_tips.groupby(['day', 'sex'])['tip'].mean().reset_index()

fig = px.bar(
    df_tips_agg,
    x='day',
    y='tip',
    color='sex',
    barmode='group',          # Barras lado a lado
    title='Propina promedio por día de la semana y género',
    labels={'tip': 'Propina promedio (USD)', 'day': 'Día', 'sex': 'Género'},
    text_auto='.2f',          # Muestra el valor sobre cada barra
    color_discrete_map={'Male': '#636EFA', 'Female': '#EF553B'}
)

fig.show()

# Barras apiladas: Población total por continente y año
df_pop = df_gapminder.groupby(['year', 'continent'])['pop'].sum().reset_index()

fig2 = px.bar(
    df_pop,
    x='year',
    y='pop',
    color='continent',
    barmode='stack',          # Barras apiladas
    title='Población mundial por continente (1952–2007)',
    labels={'pop': 'Población Total', 'year': 'Año', 'continent': 'Continente'}
)

fig2.show()

# Barras horizontales
df_pib_2007 = df_2007.nlargest(10, 'gdpPercap')

fig3 = px.bar(
    df_pib_2007,
    x='gdpPercap',
    y='country',
    orientation='h',          # Horizontal
    color='continent',
    title='Top 10 países con mayor PIB per cápita (2007)',
    labels={'gdpPercap': 'PIB per cápita (USD)', 'country': 'País'}
)

fig3.show()
````

## 4 - Histograma (Histogram)

### ¿Qué es?
Muestra la **distribución de frecuencia** de una variable numérica. Divide el rango de valores en intervalos (*bins*) y cuenta cuántas observaciones caen en cada uno.

### Conceptos clave
- **nbins**: Número de intervalos (afecta el nivel de detalle)
- **histnorm**: Normalización (`'percent'`, `'probability'`, `'density'`)
- Útil para detectar **asimetría** y **valores atípicos**

````python
# Distribución de propinas
fig = px.histogram(
    df_tips,
    x='tip',
    nbins=20,
    color='sex',
    barmode='overlay',        # Superpone los histogramas
    opacity=0.7,              # Transparencia
    marginal='rug',           # Pequeños trazos debajo mostrando cada dato
    title='Distribución de propinas por género',
    labels={'tip': 'Propina (USD)', 'sex': 'Género', 'count': 'Frecuencia'}
)

fig.show()

# Distribución de esperanza de vida por continente
fig2 = px.histogram(
    df_2007,
    x='lifeExp',
    color='continent',
    nbins=25,
    histnorm='percent',       # Normalizado como porcentaje
    marginal='box',           # Boxplot encima
    title='Distribución de Esperanza de Vida por Continente (2007)',
    labels={'lifeExp': 'Esperanza de Vida (años)', 'percent': '% de países'}
)

fig2.show()
````

---
## 5 - Box Plot (Diagrama de Caja y Bigote)

### ¿Qué es?
Resume la distribución de datos mostrando:
- **Mediana** (línea central)
- **Q1 y Q3** (bordes de la caja → rango intercuartílico)
- **Bigotes** (mínimo y máximo sin outliers)
- **Outliers** (puntos fuera de los bigotes)

Es ideal para **comparar distribuciones** entre grupos.

````python
# Box plot de propinas por día y género
fig = px.box(
    df_tips,
    x='day',
    y='tip',
    color='sex',
    notched=True,             # Muesca indica IC de la mediana
    points='outliers',        # Muestra solo los outliers
    title='Distribución de propinas por día y género',
    labels={'tip': 'Propina (USD)', 'day': 'Día', 'sex': 'Género'}
)

fig.show()

# Box plot de esperanza de vida por continente
fig2 = px.box(
    df_gapminder,
    x='continent',
    y='lifeExp',
    color='continent',
    points='all',             # Muestra TODOS los puntos
    title='Distribución de Esperanza de Vida por Continente (todos los años)',
    labels={'lifeExp': 'Esperanza de Vida (años)', 'continent': 'Continente'}
)

fig2.show()
````

---
## 6 - Violin Plot

### ¿Qué es?
Combina un **box plot** con una **estimación de densidad** (KDE). La forma del violín muestra dónde se concentran más los datos. Revela la forma de la distribución de forma más detallada que el box plot.

````python
# Violin plot de propinas por tipo de comida
fig = px.violin(
    df_tips,
    x='time',
    y='tip',
    color='sex',
    box=True,                 # Incluye el box plot dentro del violín
    points='all',             # Muestra todos los puntos
    violinmode='overlay',     # Superpone los violines por género
    title='Distribución de propinas: Comida vs Cena',
    labels={'tip': 'Propina (USD)', 'time': 'Servicio', 'sex': 'Género'}
)

fig.show()

# Comparar distribuciones de longitud de pétalo en flores iris
fig2 = px.violin(
    df_iris,
    x='species',
    y='petal_length',
    color='species',
    box=True,
    points='outliers',
    title='Longitud del pétalo por especie de flor Iris',
    labels={'petal_length': 'Longitud del Pétalo (cm)', 'species': 'Especie'}
)

fig2.show()
````

## 7 - Gráfica de Pastel y Dona (Pie / Donut Chart)

### ¿Qué es?
Muestra cómo se divide un **total** entre categorías. Cada rebanada representa la **proporción** de una categoría respecto al todo.

### Buenas prácticas
- Úsala solo con **pocas categorías** (máximo 6-7)
- Para muchas categorías, prefiere una barra apilada
- La gráfica de dona es más fácil de leer

````python
# Proporción de fumadores por tipo de día
df_smokers = df_tips.groupby('smoker').size().reset_index(name='count')

fig = px.pie(
    df_smokers,
    values='count',
    names='smoker',
    title='Proporción de fumadores en el restaurante',
    hole=0.4,                 # hole > 0 convierte en dona
    color_discrete_sequence=px.colors.qualitative.Set2
)

fig.update_traces(
    textposition='inside',
    textinfo='percent+label'  # Muestra % y etiqueta
)

fig.show()

# Población por continente en 2007
df_cont = df_2007.groupby('continent')['pop'].sum().reset_index()

fig2 = px.pie(
    df_cont,
    values='pop',
    names='continent',
    title='Distribución de la población mundial por continente (2007)',
    hole=0.3
)

fig2.update_traces(
    textposition='outside',
    textinfo='percent+label'
)

fig2.show()
````

## 8 - Mapa de Calor (Heatmap)

### ¿Qué es?
Representa valores en una **matriz** usando color. Es especialmente útil para visualizar:
- **Matrices de correlación** entre variables
- **Tablas de contingencia** (variables cruzadas)
- Patrones en datos bidimensionales

````python
# Matriz de correlación del dataset Iris
corr_matrix = df_iris.drop('species', axis=1).corr().round(2)

fig = px.imshow(
    corr_matrix,
    text_auto=True,           # Muestra el valor en cada celda
    color_continuous_scale='RdBu_r',   # Rojo = positivo, Azul = negativo
    zmin=-1, zmax=1,          # Escala de -1 a 1
    title='Matriz de Correlación — Dataset Iris',
    aspect='auto'
)

fig.show()

# Heatmap: Propina promedio por día y tipo de cliente
df_pivot = df_tips.pivot_table(
    values='tip',
    index='day',
    columns='time',
    aggfunc='mean'
).round(2)

fig2 = px.imshow(
    df_pivot,
    text_auto=True,
    color_continuous_scale='Viridis',
    title='Propina promedio (USD) por Día y Tipo de Servicio'
)

fig2.show()
````

---
## 9 - Gráfica de Área (Area Chart)

### ¿Qué es?
Similar a la gráfica de líneas pero con el área bajo la curva rellena. La versión apilada es útil para mostrar cómo **partes contribuyen al total** a lo largo del tiempo.

````python
# Área apilada: Población por continente a lo largo del tiempo
df_area = df_gapminder.groupby(['year', 'continent'])['pop'].sum().reset_index()

fig = px.area(
    df_area,
    x='year',
    y='pop',
    color='continent',
    title='Evolución de la Población Mundial por Continente',
    labels={'pop': 'Población', 'year': 'Año', 'continent': 'Continente'},
    line_group='continent'    # Agrupa por continente
)

fig.show()

# Área normalizada (proporciones)
fig2 = px.area(
    df_area,
    x='year',
    y='pop',
    color='continent',
    groupnorm='percent',      # Normaliza a 100%
    title='Proporción de Población Mundial por Continente (%)',
    labels={'pop': 'Porcentaje (%)', 'year': 'Año', 'continent': 'Continente'}
)

fig2.show()
````

## 10 - Scatter Matrix (SPLOM)

### ¿Qué es?
Muestra **todas las combinaciones posibles** de gráficas de dispersión entre variables numéricas en una sola figura. Es excelente para una exploración inicial de los datos (*Exploratory Data Analysis, EDA*).

**SPLOM** = *Scatter PLOt Matrix*

````python
# Scatter matrix del dataset Iris
fig = px.scatter_matrix(
    df_iris,
    dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
    color='species',          # Color por especie
    symbol='species',         # Símbolo por especie
    title='Scatter Matrix — Dataset Iris',
    labels={
        'sepal_length': 'Long. Sépalo',
        'sepal_width': 'Ancho Sépalo',
        'petal_length': 'Long. Pétalo',
        'petal_width': 'Ancho Pétalo'
    }
)

fig.update_traces(diagonal_visible=False)  # Oculta la diagonal
fig.show()

print('\nObserva cómo las especies se separan claramente en algunas combinaciones de variables.')
````

## 11 - Gráfica 3D (3D Scatter)

### ¿Qué es?
Extiende la gráfica de dispersión a **tres dimensiones**. Útil cuando quieres explorar relaciones entre tres variables numéricas simultáneamente.

### Recomendación
Usar con moderación. Las gráficas 3D pueden ser difíciles de interpretar en estático; son más útiles cuando son interactivas.

````python
# Scatter 3D del dataset Iris
fig = px.scatter_3d(
    df_iris,
    x='sepal_length',
    y='sepal_width',
    z='petal_length',
    color='species',
    symbol='species',
    size='petal_width',
    size_max=15,
    opacity=0.8,
    title='Clasificación 3D de Flores Iris',
    labels={
        'sepal_length': 'Long. Sépalo',
        'sepal_width': 'Ancho Sépalo',
        'petal_length': 'Long. Pétalo',
        'petal_width': 'Ancho Pétalo'
    }
)

fig.show()
print('\nArrastra la gráfica para rotarla en 3D. Observa cómo las especies forman grupos separados.')
````

## 12 - Treemap

### ¿Qué es?
Muestra datos **jerárquicos** usando rectángulos anidados. El área de cada rectángulo es proporcional a su valor. Es excelente para:
- Ver proporciones en estructuras jerárquicas
- Comparar tamaños de categorías y subcategorías

````python
# Treemap: Población mundial por continente y país (2007)
fig = px.treemap(
    df_2007,
    path=[px.Constant('Mundo'), 'continent', 'country'],   # Jerarquía: Mundo > Continente > País
    values='pop',             # Tamaño del rectángulo = población
    color='lifeExp',          # Color = esperanza de vida
    color_continuous_scale='RdYlGn',
    hover_data={'gdpPercap': ':.0f'},
    title='Población y Esperanza de Vida por País (2007)',
    color_continuous_midpoint=df_2007['lifeExp'].mean()
)

fig.update_traces(textinfo='label+percent entry')
fig.show()

print('\nHaz clic en un continente para hacer zoom en sus países.')
````

---
## 13 - Sunburst (Diagrama de Rayos de Sol)

### ¿Qué es?
Similar al Treemap pero en formato **circular**. Cada anillo representa un nivel de la jerarquía. Es muy intuitivo para mostrar cómo las categorías padre se dividen en subcategorías.

````python
# Sunburst: Propinas por día, tipo de servicio y género
df_tips_agg2 = df_tips.groupby(['day', 'time', 'sex'])['tip'].sum().reset_index()

fig = px.sunburst(
    df_tips_agg2,
    path=['day', 'time', 'sex'],   # Jerarquía
    values='tip',
    color='time',
    title='Propinas totales por Día > Servicio > Género',
    color_discrete_map={'Dinner': '#FFA15A', 'Lunch': '#19D3F3'}
)

fig.show()
print('\nHaz clic en los segmentos para navegar por la jerarquía.')
````

## 14 - Mapas Geográficos (Choropleth Maps)

### ¿Qué es?
Visualiza datos sobre un **mapa geográfico**. Los países o regiones se colorean según el valor de una variable. Ideal para datos con dimensión geográfica.

### Tipos de mapas
- `px.choropleth` → Mapa de países/estados coloreados
- `px.scatter_geo` → Puntos sobre mapa
- `px.choropleth_mapbox` → Mapa interactivo con Mapbox

````python
# Mapa coroplético: Esperanza de vida por país (2007)
fig = px.choropleth(
    df_2007,
    locations='iso_alpha',        # Código ISO del país
    color='lifeExp',              # Variable a colorear
    hover_name='country',
    color_continuous_scale=px.colors.sequential.Plasma,
    projection='natural earth',
    title='Esperanza de Vida por País (2007)',
    labels={'lifeExp': 'Esperanza de Vida (años)'}
)

fig.show()

# Mapa de PIB per cápita
fig2 = px.choropleth(
    df_2007,
    locations='iso_alpha',
    color='gdpPercap',
    hover_name='country',
    hover_data={'pop': ':,.0f'},
    color_continuous_scale=px.colors.sequential.YlOrRd,
    projection='orthographic',    # Proyección esférica
    title='PIB per Cápita por País (2007)',
    labels={'gdpPercap': 'PIB per cápita (USD)'}
)

fig2.show()
print('\nEn la proyección ortográfica puedes rotar el globo terráqueo.')
````

## 15 - Animaciones

### ¿Qué es?
Plotly Express permite crear gráficas **animadas** que muestran cómo cambian los datos a lo largo del tiempo. Basta con especificar el parámetro `animation_frame` para que la gráfica incluya controles de reproducción.

Esta es una de las funcionalidades más poderosas y atractivas de Plotly Express.

````python
# La famosa animación de Hans Rosling: PIB vs Esperanza de Vida animada por año
fig = px.scatter(
    df_gapminder,
    x='gdpPercap',
    y='lifeExp',
    animation_frame='year',       # ← ¡Este es el parámetro de animación!
    animation_group='country',    # Identifica cada punto a través del tiempo
    color='continent',
    size='pop',
    hover_name='country',
    size_max=55,
    log_x=True,
    range_x=[200, 100000],
    range_y=[25, 90],
    title='Evolución del PIB y Esperanza de Vida (1952–2007)',
    labels={
        'gdpPercap': 'PIB per cápita (USD, escala log)',
        'lifeExp': 'Esperanza de Vida (años)',
        'continent': 'Continente',
        'pop': 'Población'
    }
)

fig.show()
print('\n▶️ Presiona el botón PLAY para ver la animación.')
print('   Observa cómo el mundo se ha vuelto más rico y saludable desde 1952.')

# Animación en mapa: Esperanza de vida por país a través del tiempo
fig2 = px.choropleth(
    df_gapminder,
    locations='iso_alpha',
    color='lifeExp',
    hover_name='country',
    animation_frame='year',
    color_continuous_scale=px.colors.sequential.Plasma,
    range_color=[25, 90],
    title='Esperanza de Vida Mundial Animada (1952–2007)',
    labels={'lifeExp': 'Esperanza de Vida (años)'}
)

fig2.show()
````

## 16 - Facets (Sub-gráficas)

### ¿Qué es?
Los **facets** dividen automáticamente una gráfica en múltiples sub-gráficas según los valores de una variable categórica. Permiten comparar fácilmente la misma gráfica entre grupos.

### Parámetros
- `facet_col` → Sub-gráficas en columnas
- `facet_row` → Sub-gráficas en filas
- `facet_col_wrap` → Número máximo de columnas

````python
# Histograma por tipo de día y fumador/no fumador
fig = px.histogram(
    df_tips,
    x='total_bill',
    color='sex',
    facet_col='time',             # Sub-gráficas por tipo de servicio
    facet_row='smoker',           # Sub-gráficas por fumador
    nbins=20,
    title='Distribución de la cuenta total (por servicio y hábito de fumar)',
    labels={'total_bill': 'Cuenta total (USD)', 'sex': 'Género'}
)

fig.show()

# Scatter con facets por continente
fig2 = px.scatter(
    df_gapminder[df_gapminder['year'] == 2007],
    x='gdpPercap',
    y='lifeExp',
    color='continent',
    size='pop',
    hover_name='country',
    facet_col='continent',
    facet_col_wrap=3,             # Máximo 3 columnas
    log_x=True,
    title='PIB vs Esperanza de Vida por Continente (2007)',
    labels={'gdpPercap': 'PIB per cápita (log)', 'lifeExp': 'Esp. Vida (años)'}
)

fig.update_yaxes(matches=None)    # Escala independiente por sub-gráfica
fig2.show()
````

## 17 - Personalización: Temas, Colores y Estilos

### ¿Por qué personalizar?
Una gráfica bien diseñada comunica mejor. Plotly Express ofrece múltiples opciones para adaptar el aspecto visual a tu presentación o publicación.

### Temas disponibles
`plotly`, `plotly_white`, `plotly_dark`, `ggplot2`, `seaborn`, `simple_white`, `presentation`, `xgridoff`, `ygridoff`, `none`
